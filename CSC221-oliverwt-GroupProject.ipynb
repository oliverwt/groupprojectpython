```python
# Import CSV
import csv

# INSERT SSL FIX CODE HERE
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

# Import the library to open URLs
from urllib.request import Request, urlopen

# Import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup

# Specify which URL/web page we are going to be scraping
url = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population"
headers = {'User-Agent': 'Mozilla/5.0'}

# Open the url using urllib.request and put the HTML into the page variable
# page = urllib.request.urlopen(url)
req = Request(url, headers=headers)
page = urlopen(req).read()

# Parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "html.parser")

# Take a look at the underlying HTML
print(BeautifulSoup.prettify(soup))

# Get Table
tables = soup.find_all('table', {'class': 'wikitable'})
main_table = tables[0]
rows = main_table.find_all('tr')

with open ('statepopulations.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    for row in rows:
        # Get all header and data cells
        cols = row.find_all(['th', 'td'])

        # Extract text from each cell
        cols = [col.get_text(strip=True) for col in cols]

        # Write row to CSV (skip empty rows if you want)
        if cols:
            writer.writerow(cols)
            
import matplotlib.pyplot as plt

states = []
populations = []

with open('statepopulations.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header row
    
    for row in reader:
        if len(row) >= 2:
            state_name = row[0]
            try:
                pop_str = row[1].replace(',', '')
                population = int(pop_str)
                states.append(state_name)
                populations.append(population)
            except (ValueError, IndexError):
                continue

# Create bar graph for top 10 states
top_n = 10
top_states = states[:top_n]
top_populations = populations[:top_n]

plt.figure(figsize=(12, 6))
plt.bar(top_states, top_populations, color='steelblue', edgecolor='black')

plt.xlabel('States', fontsize=12, fontweight='bold')
plt.ylabel('Population', fontsize=12, fontweight='bold')
plt.title(f'Top {top_n} Most Populous U.S. States', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')

for i, (state, pop) in enumerate(zip(top_states, top_populations)):
    plt.text(i, pop + 500000, f'{pop:,}', ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('state_population_bargraph.png', dpi=300)
print("Bar graph saved as 'state_population_bargraph.png'")
plt.show()

print(f"\nTop 10 States by Population:")
for i, (state, pop) in enumerate(zip(top_states, top_populations), 1):
    print(f"{i}. {state}: {pop:,}")
```
